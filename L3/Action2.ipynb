```python
# -*- coding:utf-8 -*-
# 词云展示
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from lxml import etree
from nltk.tokenize import word_tokenize
```

```python
data=pd.read_csv("./Market_Basket_Optimisation.csv",header=None,sep=',')
```

```python
content=[]
```

```python
for i in data.values:
    for j in i:
        if str(j) is not None and str(j)!='nan':
            content.append(str(j).strip())
        else:
            pass
```

```python
content
```

```python
# 生成词云
def create_word_cloud(input_text):
	print('根据词频，开始生成词云!')
	wc = WordCloud(
		max_words=10,
		width=2000,
		height=1200,
    )
	wordcloud = wc.generate(input_text)
	# 写词云图片
	wordcloud.to_file("wordcloud.jpg")
	# 显示词云文件
	plt.imshow(wordcloud)
	plt.axis("off")
	plt.show()
```

```python

# 生成词云
create_word_cloud(",".join(content))
```
