```python
#针对MarketBasket数据集进行购物篮分析（频繁项集及关联规则挖掘）
```

```python
import pandas as pd
from efficient_apriori import apriori
```

```python
data=pd.read_csv("./Market_Basket_Optimisation.csv",header=None,sep='|')
```

```python
data[0]=data[0].str.lower()
```

```python
data=data[0].str.get_dummies(',') # It will auto remove duplicated onehot attribute
```

```python
data.columns.tolist()
```

```python
transactions=[]
```

```python
columns=data.columns.tolist()
```

```python
columns
```

```python
def prepare_data(row):
    temp_set=set()
    for k,v in enumerate(row):
        if v == 1 :
            temp_set.add(columns[k])
        else :
            pass
        
    transactions.append(tuple(temp_set))
    #print("+++")
```

```python
data.apply(lambda row: prepare_data(row),axis=1)
```

```python
transactions
```

```python
# 挖掘频繁项集和频繁规则
itemsets, rules = apriori(transactions, min_support=0.01,  min_confidence=0.5)
print('频繁项集：', itemsets)
print('关联规则：', rules)
```
